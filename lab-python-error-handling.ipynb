````python

# Define functions

def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

def get_customer_orders(inventory):
    print("\n\n")
    valid_num_orders = False
    while not valid_num_orders:
        try:
            num_orders = int(input("Enter the number of customer orders: "))
            if num_orders < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
            valid_num_orders = True
        except ValueError:
            print("Please enter a valid number of orders.")
    
    orders = []
    for i in range(num_orders):
        valid_product = False
        while not valid_product:
            try:
                order = input(f"Enter the name of a product:")
                if inventory[order] == 0:
                    print("Sorry, that product is out of stock. Please enter a different product.")
                    valid_product = False
                elif order in inventory:
                    valid_product = True
                    orders.append(order)
                else:
                    print("Please enter a valid product name.")
            except :
                print("Please enter a valid product name.") 

    return orders

def calculate_total_price(orders):
    valid_order = False
    while not valid_order:
        try:
            total_price = sum([float(input(f"Enter the price of {order}: ")) for order in orders])
            valid_order = True
        except ValueError:
            print("Please enter a valid price.")
    return total_price

def update_inventory(inventory, orders):
    for order in orders:
        if order in inventory:
            inventory[order] -= 1
    inventory = {product: quantity for product, quantity in inventory.items() if quantity > 0}
    return inventory

def print_total_price(total_price):
    print("\n\n")
    print(f"Total Price: {total_price}")

def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    total_products_available = len(products)
    percentage_ordered = total_products_ordered / total_products_available * 100
    return (total_products_ordered, percentage_ordered)

def print_order_statistics(order_statistics):
    print("\n\nOrder Statistics:")
    print("Total Products Ordered:", order_statistics[0])
    print("Percentage of Unique Products Ordered:", order_statistics[1])




# Initialization
products = ["t-shirt", "mug", "hat", "book", "keychain"]

# Start
inventory = initialize_inventory(products)
customer_orders = get_customer_orders(inventory)


order_statistics = calculate_order_statistics(customer_orders, products)
print_order_statistics(order_statistics)

updated_inventory = update_inventory(inventory, customer_orders)


print("\nUpdated Inventory:")
for product, quantity in updated_inventory.items():
    print(f"{product}: {quantity}")

total_price = calculate_total_price(customer_orders)
print_total_price(total_price)

````
